#### 1. Simulación del movimiento browniano

El movimiento browniano es el movimiento aleatorio que se observa en las partículas que se hallan en un medio fluido, como resultado de choques contra las moléculas de dicho fluido [(referencia)].

[(referencia)]: https://es.wikipedia.org/wiki/Movimiento_browniano

Para simular el movimiento, en lugar de usar las ecuaciones integro-diferenciales que describe Einstein, use una caminta aleatoria para los tres ejes $x$, $y$, $z$ de la siguiente forma
$$
W_i=W_{i-1}+X_i\sqrt{\Delta t}
$$
donde, $\Delta t$ es un intervalo de tiempo y 
$$
X_i=\left\{\begin{array}{l}
1\\
-1
\end{array}\right.,
$$
se escoge de forma aleatoria con la misma probabilidad.

Simule $10$ segundos con un $\Delta t=0.001$ y represéntelo en un gráfico 3D (Puede ver el archivo de Ejemplo 3D).

```python
import numpy as np
import random 
import sympy as sp

lista_de_opciones = ["1", "-1"]
sp.var("x")
sp.var("y")
sp.var("z")
random.choice(lista_de_opciones)
```

```python
raiz_AT = np.sqrt(0.001)
raiz_AT 
```

```python
x=[]
y=[]
z=[]
AT = 0.001
contador = 0

while contador < 10/0.001 :
    x_1 = float(random.choice(lista_de_opciones))
    y_1 = float(random.choice(lista_de_opciones))
    z_1 = float(random.choice(lista_de_opciones))
    w_1 = 0
    w_2 = 0
    w_3 = 0
    w_1 = w_1 + np.asarray(x_1)*(raiz_AT)
    w_2 = w_2 + np.asarray(y_1)*(raiz_AT)
    w_3 = w_3 + np.asarray(z_1)*(raiz_AT)
    x_num= x.append(w_1)
    y_num=y.append(w_2)
    z_num=z.append(w_3)
    contador = contador + 1
x
```

#### 2. Apuesta arriesgada

Un jugador, con 20 pesos de capital inicial, decide apostar a un volado (con una moneda perfectamente nivelada) la mitad de su capital si éste es mayor o igual a 6 pesos, de lo contrario, solo 2. Esto es, el capital $C_i$ en cada volado $V_i$ se describe como
$$
C_{i+1}=C_i+V_iA_i
$$
donde la apuesta es 
$$
A_i=\left\{\begin{array}{l}
C_i/2,\quad \mbox{si }C_i\geq 6\\
2,\quad \mbox{de lo contrario.}
\end{array}\right.
$$

1. Grafique la evolución del capital hasta quedar en cero.
2. Determine el número de volados promedio para quedar en quiebra usando la simulación montecarlo para 10,000 apostadores.
3. Muestre un histograma del número de volados.

# Grafique la evolución del capital hasta quedar en cero

```python
import numpy as np
import random 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def capital():
    return (20)
capital()
```

```python
evolucion = [capital]
```

```python
while capital > 0:
    if capital >= 6:
        apuesta = capital/2
    else:
        apuesta = 2
```

```python
volado = random.randint(0,1)
if volado == 0:
        capital = capital - apuesta
else:
        capital = capital + apuesta
        
evolucion.append(capital)
```

```python
plt.plot(evolucion)
plt.xlabel("VOLADOS")
plt.ylabel("CAPITAL")
plt.title('Evolución del capital hasta quedar en cero')
plt.show()
```

# Determine el número de volados promedio para quedar en quiebra usando la simulación montecarlo para 10,000 apostadores.

```python
def jugar_hasta_quiebra(capital):
    num_volados = 0
    while capital > 0:
        if capital >= 6:
            apuesta = capital/2
        else:
            apuesta = 2
            
        resultado = random.randint(0,1)
        if resultado == 0:
            capital = capital - apuesta
        else:
            capital = capital + apuesta
            
        num_volados += 1
        
    return num_volados

num_jugadores = 10000
num_volados_promedio = 0

for i in range(num_jugadores):
    num_volados_jugador = jugar_hasta_quiebra(20)
    num_volados_promedio += num_volados_jugador
```

```python
promedio = num_volados_promedio / num_jugadores
promedio
```

# Muestre un histograma del número de volados.

```python
num_jugadores = 10000
num_volados = []

for i in range(num_jugadores):
    num_volados_jugador = jugar_hasta_quiebra(20)
    num_volados.append(num_volados_jugador)
    
plt.hist(num_volados, bins=20)
plt.xlabel("VOLADOS")
plt.ylabel("JUGADORES")
plt.title("VOLADOS PROMEDIO PARA QUEDAR EN QUIEBRA")
plt.show()
```
